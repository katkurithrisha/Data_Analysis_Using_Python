1. Create Vectors, Factors, Lists, Matrices, and Data Frames

```python
import pandas as pd
file_path = '/content/insurance.csv'
insurance_data = pd.read_csv(file_path)
# Vector
ages = insurance_data['age']
smokers = insurance_data['smoker']
expenses = insurance_data['expenses']
print("Vector - Ages:\n", ages.head(), "\n")
print("Vector - Smokers:\n", smokers.head(), "\n")
print("Vector - Expenses:\n", expenses.head(), "\n")
# Factor
smoker_status = insurance_data['smoker'].unique()
print("Factor - Smoker Status:\n", smoker_status, "\n")
# List
data_list = [ages, smokers, expenses]
print("List of Data:\n", [type(item) for item in data_list], "\n")
# Matrix
data_matrix = insurance_data[['age', 'expenses']].values
print("Matrix - Age and Expenses:\n", data_matrix[:5], "\n")
# DataFrame
print("DataFrame Head:\n", insurance_data.head(), "\n")
```

Scatter Plot: Age vs Smoking

```python
import matplotlib.pyplot as plt
import seaborn as sns
file_path = '/content/insurance.csv'
insurance_data = pd.read_csv(file_path)
# Scatter Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=insurance_data, x='age', y='smoker', marker='+', color='blue')
plt.title("Scatter Plot: Age vs Smoking Status")
plt.xlabel("Age")
plt.ylabel("Smoking Status (yes/no)")
plt.grid(alpha=0.3)
plt.show()
```

Filter the Data (Age 20 to 40) and Update the Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
file_path = '/content/insurance.csv'
insurance_data = pd.read_csv(file_path)
# Filter the data for ages 20 to 40
filtered_data = insurance_data[(insurance_data['age'] >= 20) & (insurance_data['age'] <= 40)]
print("Filtered Data (Ages 20-40):\n", filtered_data.head(), "\n")
# Scatter Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=filtered_data, x='age', y='smoker', marker='o', color='green')
plt.title("Scatter Plot: Age (20-40) vs Smoking Status")
plt.xlabel("Age")
plt.ylabel("Smoking Status (yes/no)")
plt.grid(alpha=0.3)
plt.show()
```

Plot Total Expenses: Smokers vs Non-Smokers



```python
import pandas as pd
import matplotlib.pyplot as plt
file_path = '/content/insurance.csv'
insurance_data = pd.read_csv(file_path)
# Group data by smoker status and calculate total expenses
expenses_grouped = insurance_data.groupby('smoker')['expenses'].sum()
print("Total Expenses (Smokers vs Non-Smokers):\n", expenses_grouped, "\n")
# Bar Plot
expenses_grouped.plot(kind='bar', color=['orange', 'purple'], figsize=(8, 6))
plt.title("Total Expenses: Smokers vs Non-Smokers")
plt.ylabel("Total Expenses")
plt.xlabel("Smoking Status")
plt.xticks(rotation=0)
plt.show()
```

Find the Number of Smokers Gender-Wise

```python
import pandas as pd
import matplotlib.pyplot as plt
file_path = '/content/insurance.csv'
insurance_data = pd.read_csv(file_path)
# Count the number of smokers by gender
smokers_gender = insurance_data[insurance_data['smoker'] == 'yes'].groupby('sex').size()
print("Number of Smokers by Gender:\n", smokers_gender, "\n")
# Bar Plot
smokers_gender.plot(kind='bar', color=['red', 'blue'], figsize=(8, 6))
plt.title("Number of Smokers by Gender")
plt.ylabel("Number of Smokers")
plt.xlabel("Gender")
plt.xticks(rotation=0)
plt.show()
```

Scatter Plot: Age vs Expenses

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
file_path = '/content/insurance.csv'
insurance_data = pd.read_csv(file_path)
# Scatter Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=insurance_data, x='age', y='expenses', hue='smoker', style='sex', palette='coolwarm', s=60)
plt.title("Scatter Plot: Age vs Expenses")
plt.xlabel("Age")
plt.ylabel("Expenses")
plt.grid(alpha=0.3)
plt.legend(title="Legend", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
